```python
# create mnists digits image
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load data from https://www.openml.org/d/554
X, y = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)
```

```python
# parameter: how many columns do you want
cols = 12

# create database selection
digs = []
for j in range(0, 10):
    zs = []
    for i, v in enumerate(y):
        if v == str(j):
            zs.append(i)
            if len(zs) == cols: 
                digs.append(zs)
                break
# plot
# parameter: scaling factor
sk = 0.8
# parameter: padding factor
pd = 0.05
# parameter: color map (pink, gray, magma, inferno, cividis, ocean)
cm = 'magma'
fig, axes = plt.subplots(10, cols, figsize=(sk*cols, 10*sk))
for j in range(0, 10):
    for i in range(0, cols):
        axes[j][i].get_yaxis().set_visible(False)
        axes[j][i].get_xaxis().set_visible(False)
        axes[j][i].set_xticklabels([])
        axes[j][i].set_yticklabels([])
        axes[j][i].set_aspect('equal')
        axes[j][i].imshow(X[digs[j][i]].reshape(28,28), cmap=cm)
plt.subplots_adjust(wspace=pd*sk, hspace=pd*sk)
# plt.show()
plt.savefig('mnist.png')
```
